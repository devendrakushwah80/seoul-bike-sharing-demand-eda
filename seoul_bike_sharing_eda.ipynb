```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True),annot=True, fmt=".2f",cmap='coolwarm')
```

```python
plt.bar(df['Temperature(°C)'], df['Rented Bike Count'])
plt.xlabel("Temprature")
plt.ylabel("Total Rentals")
plt.title("Total Bike Rentals vs Temprature")
plt.show()
```

```python
plt.bar(df['Hour'], df['Rented Bike Count'])
plt.xlabel("Hour")
plt.ylabel("Total Rentals")
plt.title("Total Bike Rentals Per Hour")
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'],dayfirst=True)
```

```python
plt.bar(df['Date'], df['Rented Bike Count'])
plt.xlabel("Date")
plt.ylabel("Total Rentals")
plt.title("Total Bike Rentals Per Date")
plt.show()
```

```python
plt.bar(df['Dew point temperature(°C)'], df['Rented Bike Count'])
plt.xlabel("Dew point temperature(°C)")
plt.ylabel("Total Rentals")
plt.title("Total Bike Rentals Per Dew point temperature(°C)")
plt.show()
```

```python
plt.plot(df['Dew point temperature(°C)'], df['Temperature(°C)'])
plt.xlabel("Dew point temperature(°C)")
plt.ylabel("Temperature(°C)")
plt.title("Dew point temperature(°C) per Temperature(°C))")
plt.show()
```

```python
sns.boxplot(x=df['Holiday'], y=df['Rented Bike Count'])
plt.xlabel("Holiday")
plt.ylabel("Total Rentals") 
plt.title("Total Bike Rentals on Holidays vs Non-Holidays")
plt.show()
```

```python
holiday_counts = df.groupby("Holiday")["Rented Bike Count"].sum()

plt.figure(figsize=(7,7))
plt.pie(holiday_counts, labels=holiday_counts.index, autopct="%1.1f%%")
plt.title("Bike Rentals: Holiday vs Non-Holiday")
plt.show()
```

```python
sns.boxplot(x=df['Seasons'], y=df['Rented Bike Count'])
plt.xlabel("Seasons")
plt.ylabel("Total Rentals") 
plt.title("Total Bike Rentals in Different Seasons")   
plt.show()  
```

```python
season_counts = df.groupby("Seasons")["Rented Bike Count"].sum()
plt.figure(figsize=(8,8))
plt.pie(season_counts, labels=season_counts.index, autopct="%1.1f%%")
plt.title("Total Bike Rentals by Season")
plt.show()
```
