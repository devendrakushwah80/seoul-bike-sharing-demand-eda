```python
plt.bar(df['Temperature(°C)'], df['Rented Bike Count'])
plt.xlabel("Temprature")
plt.ylabel("Total Rentals")
plt.title("Total Bike Rentals vs Temprature")
plt.show()
```

```python
plt.bar(df['Hour'], df['Rented Bike Count'])
plt.xlabel("Hour")
plt.ylabel("Total Rentals")
plt.title("Total Bike Rentals Per Hour")
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'],dayfirst=True)
```

```python
plt.bar(df['Date'], df['Rented Bike Count'])
plt.xlabel("Date")
plt.ylabel("Total Rentals")
plt.title("Total Bike Rentals Per Date")
plt.show()
```

```python
plt.bar(df['Dew point temperature(°C)'], df['Rented Bike Count'])
plt.xlabel("Dew point temperature(°C)")
plt.ylabel("Total Rentals")
plt.title("Total Bike Rentals Per Dew point temperature(°C)")
plt.show()
```
