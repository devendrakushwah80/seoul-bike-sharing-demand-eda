```python
plt.bar(df['Temperature(°C)'], df['Rented Bike Count'])
plt.xlabel("Temprature")
plt.ylabel("Total Rentals")
plt.title("Total Bike Rentals vs Temprature")
plt.show()
```

## We CAn Increase Bike Supply During Temperature between 20 and 30 degree or charge more

```python
plt.bar(df['Hour'], df['Rented Bike Count'])
plt.xlabel("Hour")
plt.ylabel("Total Rentals")
plt.title("Total Bike Rentals Per Hour")
plt.show()
```

## we can decrease price between 0 to 10 hours due to less number of bike rented and also we can increase price in peek hours between 10 to 20

```python
df['Date'] = pd.to_datetime(df['Date'],dayfirst=True)
```

```python
plt.bar(df['Date'], df['Rented Bike Count'])
plt.xlabel("Date")
plt.ylabel("Total Rentals")
plt.title("Total Bike Rentals Per Date")
plt.show()
```

```python
plt.bar(df['Dew point temperature(°C)'], df['Rented Bike Count'])
plt.xlabel("Dew point temperature(°C)")
plt.ylabel("Total Rentals")
plt.title("Total Bike Rentals Per Dew point temperature(°C)")
plt.show()
```

## We can increase price when dew point is between -10 to 20 and decrease price between -30 to -10

```python
plt.plot(df['Dew point temperature(°C)'], df['Temperature(°C)'])
plt.xlabel("Dew point temperature(°C)")
plt.ylabel("Temperature(°C)")
plt.title("Dew point temperature(°C) per Temperature(°C))")
plt.show()
```

```python
sns.boxplot(x=df['Holiday'], y=df['Rented Bike Count'])
plt.xlabel("Holiday")
plt.ylabel("Total Rentals") 
plt.title("Total Bike Rentals on Holidays vs Non-Holidays")
plt.show()
```

## We can give subscription or monthly pass for holidays with less price 

```python
holiday_counts = df.groupby("Holiday")["Rented Bike Count"].sum()

plt.figure(figsize=(7,7))
plt.pie(holiday_counts, labels=holiday_counts.index, autopct="%1.1f%%")
plt.title("Bike Rentals: Holiday vs Non-Holiday")
plt.show()
```

```python
sns.boxplot(x=df['Seasons'], y=df['Rented Bike Count'])
plt.xlabel("Seasons")
plt.ylabel("Total Rentals") 
plt.title("Total Bike Rentals in Different Seasons")   
plt.show()  
```

```python
season_counts = df.groupby("Seasons")["Rented Bike Count"].sum()
plt.pie(season_counts, labels=season_counts.index, autopct="%1.1f%%")
plt.title("Total Bike Rentals by Season")
plt.show()
```

## we can increase prices during peak hours and sunny days and also we can increase bike numbers and sale is less in winter so we can provide winter gears and warm seat cover and gloves in all bikes

```python
sns.scatterplot(x='Snowfall (cm)', y='Rented Bike Count', data=df)
plt.title("Snowfall vs Bike Rentals")
plt.xlabel("Snowfall (cm)")
plt.ylabel("Total Rentals")
plt.show()
```

## We can see when there is no snowfall high number of bike rented so we can provide winter ready bikes like winter tyres or weather protection gears

```python
sns.scatterplot(x='Rainfall(mm)', y='Rented Bike Count', data=df)
plt.title("Rainfall vs Bike Rentals")
plt.xlabel("Rainfall (mm)")
plt.ylabel("Total Rentals")
plt.show()
```

## we can give discount on rainy days or provide raincoat with bike with to cover rent 

```python
sns.scatterplot(x='Visibility (10m)', y='Rented Bike Count', data=df)
plt.title("Visibility vs Bike Rentals")
plt.xlabel("Visibility (10m)")
plt.ylabel("Total Rentals")
plt.show()
```

## we can provide reflective gear for night riders to enhance their visibility and safety

```python
sns.scatterplot(x='Humidity(%)', y='Rented Bike Count', data=df)
plt.title("Bike Rentals vs Humidity")
plt.xlabel("Humidity (%)")
plt.ylabel("Total Rentals")
plt.show()
```
